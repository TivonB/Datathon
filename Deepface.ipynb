```python
from deepface import DeepFace
import csv
import os
```

```python
#Analyzes facial features and expressions in an image using DeepFace library.
def get_info(img,act):
    dfs = DeepFace.analyze(img_path = img, actions = act)
    return dfs[0]['dominant_gender'], dfs[0]['dominant_race'], dfs[0]['age']
```

```python
# List all files in the 'faceimages' directory
db_path = os.listdir('faceimages')

# Define a list of facial analysis actions to perform
action_list = ['gender', 'race', 'age']

# Open a CSV file for writing DeepFace analysis results
with open("Deepface_Result.csv", "w", newline="") as f:
    # Create a CSV writer
    writer = csv.writer(f)
    
    # Write the header row with column names
    writer.writerow(["Filename", "Gender", "Race/ethnicity", "Age"])

    # Iterate through each image in the 'faceimages' directory
    for _, img in enumerate(db_path):
        try:
            img_path = os.path.join('faceimages',img)
            # Analyze facial features using the 'get_info' function
            df_gender, df_race, df_age = get_info(img_path, action_list)
            
            # Write the analysis results to the CSV file
            writer.writerow([img, df_gender, df_race, df_age])
        except:
             # Handle exceptions by writing default values for failed analysis
            writer.writerow([img, 'unknown', 'unknown', '-1'])
```
